```python
from pyknow import *

class TaxFacts(Fact):
    pass

class TaxRules(KnowledgeEngine):
    @Rule(TaxFacts(income=lambda x: x <= 50000))
    def low_income_tax(self):
        self.declare(TaxFacts(tax_rate=0.1))

    @Rule(TaxFacts(income=lambda x: 50000 < x <= 100000))
    def medium_income_tax(self):
        self.declare(TaxFacts(tax_rate=0.15))

    @Rule(TaxFacts(income=lambda x: x > 100000))
    def high_income_tax(self):
        self.declare(TaxFacts(tax_rate=0.2))

    @Rule(TaxFacts(childcare_expenses=lambda x: x > 0))
    def childcare_deduction(self):
        self.declare(TaxFacts(deductions=lambda x, y: x + y), explanation="Deducting childcare expenses.")

    @Rule(TaxFacts(union_expenses=lambda x: x > 0))
    def union_deduction(self):
        self.declare(TaxFacts(deductions=lambda x, y: x + y), explanation="Deducting union expenses.")

    @Rule(TaxFacts(rrsp_contributions=lambda x: x > 0))
    def rrsp_deduction(self):
        self.declare(TaxFacts(deductions=lambda x, y: x + y), explanation="Deducting RRSP contributions.")

    @Rule(TaxFacts(pension_adjustment=lambda x: x > 0))
    def pension_adjustment(self):
        self.declare(TaxFacts(pension_adjustment=True), explanation="Adjusting for pension contributions.")

    @Rule(AS.f1 << TaxFacts(income=MATCH.income, tax_rate=MATCH.tax_rate),
          AS.f2 << TaxFacts(deductions=MATCH.deductions),
          AS.f3 << TaxFacts(pension_adjustment=True))
    def calculate_tax(self, f1, f2, f3):
        taxable_income = f1.income - f2.deductions
        tax = taxable_income * f1.tax_rate
        self.declare(TaxFacts(total_tax=tax))  # Store the calculated tax in a new fact

# Example usage
engine = TaxRules()

income = float(input("Enter your income: "))
childcare_expenses = float(input("Enter childcare expenses: "))
union_expenses = float(input("Enter union expenses: "))
rrsp_contributions = float(input("Enter RRSP contributions: "))
pension_adjustment = float(input("Enter pension adjustment: "))

engine.reset()

engine.declare(TaxFacts(income=income))
engine.declare(TaxFacts(childcare_expenses=childcare_expenses))
engine.declare(TaxFacts(union_expenses=union_expenses))
engine.declare(TaxFacts(rrsp_contributions=rrsp_contributions))
engine.declare(TaxFacts(pension_adjustment=pension_adjustment))

engine.run()

# Retrieve the total tax from the new fact
total_tax = engine.facts[1].total_tax

print("Total Tax: $", total_tax)
```

```python
from pyknow import *

class TaxFacts(Fact):
    pass

class TaxRules(KnowledgeEngine):
    @Rule(TaxFacts(income=P(lambda x: x <= 50000)))
    def low_income_tax(self):
        self.declare(TaxFacts(tax_rate=0.1))

    @Rule(AND(TaxFacts(income=P(lambda x: 50000 < x <= 100000)),
               TaxFacts(tax_rate=None)))  
    def medium_income_tax(self):
        self.declare(TaxFacts(tax_rate=0.15))

    @Rule(AND(TaxFacts(income=P(lambda x: x > 100000)),
              TaxFacts(tax_rate=None)))
    def high_income_tax(self):
        self.declare(TaxFacts(tax_rate=0.2))

    @Rule(TaxFacts(childcare_expenses=P(lambda x: x > 0)))
    def childcare_deduction(self):
        self.declare(TaxFacts(deductions=lambda x: x))

    @Rule(TaxFacts(union_expenses=P(lambda x: x > 0)))
    def union_deduction(self):
        self.declare(TaxFacts(deductions=lambda x: x))

    @Rule(TaxFacts(rrsp_contributions=P(lambda x: x > 0)))
    def rrsp_deduction(self):
        self.declare(TaxFacts(deductions=lambda x: x))

    @Rule(TaxFacts(pension_adjustment=P(lambda x: x > 0)))
    def pension_adjustment(self):
        self.declare(TaxFacts(pension_adjustment=True))

    @Rule(AND(TaxFacts(income=MATCH.income), 
           TaxFacts(tax_rate=MATCH.tax_rate),
           TaxFacts(deductions=MATCH.deductions),
           NOT(TaxFacts(total_tax=MATCH.total_tax))))
    def calculate_tax(self, income, tax_rate, deductions):
        taxable_income = income - deductions
        total_tax = taxable_income * tax_rate
        self.declare(TaxFacts(total_tax=total_tax))

# Example usage
engine = TaxRules()

income = float(input("Enter your income: "))
childcare_expenses = float(input("Enter childcare expenses: "))
union_expenses = float(input("Enter union expenses: "))
rrsp_contributions = float(input("Enter RRSP contributions: "))
pension_adjustment = float(input("Enter pension adjustment: "))

engine.reset()

engine.declare(TaxFacts(income=income))
engine.declare(TaxFacts(childcare_expenses=childcare_expenses))
engine.declare(TaxFacts(union_expenses=union_expenses))
engine.declare(TaxFacts(rrsp_contributions=rrsp_contributions))
engine.declare(TaxFacts(pension_adjustment=pension_adjustment))

engine.run()
```
